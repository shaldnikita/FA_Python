```python
class Node(object):
    def __init__(self,value):
        self.__children=[None,None]
        self.__value=value
    def __str__(self):
        return str(self.value)
    @property
    def leftChild(self):
        return self.__children[0]
    @leftChild.setter
    def leftChild(self,value):
        self.__children[0]=value
        
    @property
    def rightChild(self):
        return self.__children[1]
    @rightChild.setter
    def rightChild(self,value):
        self.__children[1]=value
        
    @property
    def value(self):
        return self.__value
    @value.setter
    def value(self,newValue):
        self.__value=newValue
    
    @property
    def hasChildren(self):
        return len(self.__children)!=0
    @property
    def children(self):
        return self.__children
    def getLeftChild(self):
        return BinaryTreeLL(self.leftChild)
    def getRightChild(self):
        return BinaryTreeLL(self.rightChild)
            
        
    
        
class BinaryTreeLL(object):
    def __init__(self,root=None):
        self.root=root
        
    def add(self,value):
        newNode = Node(value)
        current=self.root
        parent=current
        
        if current==None:
            self.root=newNode
            return
        else:
            while(True):
                parent=current
                #print("parent= ",type(parent))
                if(value>current.value):
                    current=parent.rightChild
                    #print("current= ",type(current))
                    if current==None:
                        parent.rightChild=newNode
                        return
                else:
                    current=parent.leftChild
                    if current==None:
                        parent.leftChild=newNode
                        return
bt = BinaryTreeLL()
bt.add(3)
bt.add(1)
bt.add(2)
```
