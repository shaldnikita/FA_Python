* get_val() - возвращает объект, хранящийся в данном узле.

* get_tree_list() - возвращает представление дерева в виде списка.

```python
class Node(object):
    def __init__(self,value,parent):
        self.parent=parent
        self.leftChild=None
        self.rightChild=None
        self.value=value
        
    def __str__(self):
        return str(self.value)
        
class BinaryTree(object):
    def __init__(self):
        self.root=None
        
    def find(self,value):
        current = self.root
        if current==None:
            return "EmptyTree"
        else:
            while(current.value!=value):
                    if value>current.value:
                        current=current.rightChild
                    else:
                        current = current.leftChild
            return current
    def add(self,value):
        current = self.root
        newNode = Node(value,current)
        parent=current
        if current==None:
            self.root=newNode
            return
        else:
            while(True):
                parent=current
                if value<current.value:
                    current=current.leftChild
                    if(current==None):
                        newNode = Node(value,parent)
                        parent.leftChild=newNode
                        return
                else:
                    current=current.rightChild
                    if(current==None):
                        newNode = Node(value,parent)
                        parent.rightChild=newNode
                        return
    @staticmethod
    def displayTree(node):
        if node==None:
            return
        #print(node)
        if node.leftChild!=None:
            print(node.leftChild)
        if node.rightChild!= None:
            print(node.rightChild)
            
        left=node.leftChild
        right=node.rightChild
        BinaryTree.displayTree(right)
        BinaryTree.displayTree(left)
        
    
bt = BinaryTree()
bt.add(3)
bt.add(4)
bt.add(1)
bt.add(2)
```
